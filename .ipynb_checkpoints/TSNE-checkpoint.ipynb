```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
from utils import KNN
rand_state = 7
```

```python
gcn = np.load("embeddings_GCN_glove.npy")
gcn_embeds = TSNE(n_components=2, random_state=rand_state).fit_transform(gcn[:,0:-1])
gcn_labels = gcn[:,-1].astype(int)
df_gcn = pd.DataFrame({'x1':gcn_embeds[:,0],'x2':gcn_embeds[:, 1], 'label':gcn_labels})
#for n in [2,5,8,10,15]:
#    KNN(gcn[:,0:-1], gcn_labels, n)
```

```python
file = "glove8-2.csv"
X = pd.read_csv("../LatentSemanticImputation/word_classification/mlr/" + file, index_col = 0).drop('y', axis = 1).values
y = pd.read_csv("../LatentSemanticImputation/word_classification/mlr/" + file, index_col = 0)['y'].values
lsi_embeds = TSNE(n_components=2, random_state= rand_state).fit_transform(X)
lsi_labels = y.astype(int)
df_lsi = pd.DataFrame({'x1':lsi_embeds[:,0],'x2':lsi_embeds[:, 1], 'label':lsi_labels})
#for n in [2,5,8,10,15]:
#    KNN(X,y,n)
```

```python
fig, ax =plt.subplots(1,2, figsize = (20,10))

sns.scatterplot(x="x1", y="x2", ax = ax[0], hue="label", data=df_gcn, s = 50,
palette=['green','orange','pink','blue','red','gray','yellow','black','brown', 'dodgerblue', 'purple'], legend='full')

sns.scatterplot(x="x1", y="x2", ax = ax[1], hue="label", data=df_lsi, s = 50,
    palette=['green','orange','pink','blue','red','gray','yellow','black','brown', 'dodgerblue', 'purple'], legend='full')

fig.show()
```
