```python
import pandas as pd
import numpy as np
```

```python
aff = pd.read_csv("./data/finance/aff.txt", index_col = 0, header = None, sep = " ")
subglove = pd.read_csv("./data/finance/subglove.txt", index_col = 0, header = None, sep = " ")
tokens = pd.read_csv("./data/finance/finance_token.csv")[['nGram','Sector']]
merged = aff.merge(tokens, how = 'left', left_on = 0, right_on = 'nGram')
```

```python
unique_sectors = sorted(merged['Sector'].unique())
map_dict = dict(enumerate(unique_sectors))
map_dict = {v:k for k,v in map_dict.items()}
merged['y'] = merged['Sector'].map(lambda x: map_dict[x])
merged['y'] = merged['y'] + 1
```

```python
company_all = merged['key_0'].to_list()
glove_index = subglove.index.to_list()
company_in_glove = [i for i in company_all if i in glove_index]
```

```python
common_labels = merged[merged.key_0.isin(company_in_glove)]['y'].astype(int).values
```

```python
common_embeds = subglove[subglove.index.isin(company_in_glove)]
common_embeds['y'] = common_labels
common_embeds['count'] = 0 # empty column to fit with the existing eval.py
uncommon_embeds = subglove[~subglove.index.isin(company_in_glove)]
uncommon_embeds.to_csv('./data/finance/gloveMat_4000_uncommon.csv')
common_embeds.to_csv('./data/finance/gloveMat_4000.csv')
```
